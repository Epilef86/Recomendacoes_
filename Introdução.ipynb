Sistemas de Recomendação veio a partir de um movimento de digitalização das lojas. As lojas nas décadas de 80, 90, em qua as características delas era de serem lojas físicas, tinham um espaço físico limitado com prateleiras, então assim vc tinha uma limitação no catálogo dos seus produtos pra ofertar pros clientes e tinha clientes centralizados, ou seja, loja focada em clientes de uma determinada região. 
Quando começou a ter a migração das lojas pro ambiente digital, em que as empresas começaram a oferecer produtos aos cliente via web, através de e-commerce. Esse espaço de oferta começou a ficar "ilimitado" onde vc conseguia oferecer pros usuários produtos via web e aí como não tinha limitação no espaço físico vc poderia oferecer milhares ou milhões de produtos e além disso, podíamos descentralizar nossos clientes. Então pode vender pro usuário ou cliente que está em um outro país. Conforme vamos aumentando a oferta dos produtos, cria-se um problema em que vc quer gerar receita pra produtos de nincho, ás vezes vc não consegue mostrar muito pro usuário todo catálogo, justamente pela gama de produtos que se oferece. Porém de acordo com a Teoria da Cauda-Loga, o autor defendia que:

"A receita gerada pro produtos de nicho pode ser tão alta quanto a receita de produtos blockbuster"

Então, pensa na empresa como a Amazon, em que ela tem em estoque alguns produtos bem ninchado e que a empresa precisa gerar receita pra aqueles produtos, a Teoria da Cauda-Loga é justamente pra isso, conseguir de forma inteligente gerar receita pra esses produtos de nicho, ou seja, produtos de baixa popularidade:

Topologias de Sistemas de Recomendações:

1- Recomendação Não Personalizadas;

2- Recomendação Content-Basead;

3- Recomendação Filtragem Colaborativa;

4- Recomendação Híbrida

1- Recomendação Não Personalizadas é caracterizada por fazer a mesma recomendação para todos os usuários, é como se fosse uma transmissão única pra todos os usuários. Geralmente utilizamos isso quando se tem um usuário que acabou de entrar em seu sistema e vc nao tem dados sobre esse usuário e ai esse usuário sofre do que chamamos de cold-start de usuário, a recomendação demora um tempo pra vc ter uma qualidade na recomendação pra aquele usuário pq não sabemos nada ainda sobre ele. 

Um outra característica em Recomendação Não Personalizadas é que ela é de rápida atualização (streaming), geralmente usamos isso em pipeline de streaming, exemplo: saber quais os filmes mais assistidos ou as matérias que estão sendo mais acessadas

Baixa Complexidade, não considera o perfil de consumo

2- Recomendação Content-Basead

Foco é nas features dos itens, ou seja, pega um usuário e ver quais são os itens que esse usuário consumiu e tenta achar itens parecidos com que esse usuário consumiu 

Vantagens: cold start de item -  assim como demora a fazer recomendação de qualidade ao um novo usuário, o item também tem um cold start, geralmente precisa que usuários suficientes consumam esse item pra esse item começar a ser recomendado. O Content-Basead não sofre disso pq eu só preciso extrair as features dos itens pra esse item começar a ser recomendado. 

Uma característica muito importante dessa recomendação é a alta explicabilidade, o usuário quando ele tende o pq aquela recomendação tá indo pra ele, ele passa a confiar muito mais na recomendação.

Uma limitação dessa recomendação é extração do features, nem sempre seu item tem uma extração fácil das features.

Outra característica não muito desejada é o efeito bolha, se está considerando que o usuário assistiu e está desconsiderando o consumo de outros usuários a tedência é de que vc mantenha esse usuário numa bolha de consumo

image.png

3- Recomendação Filtragem Colaborativa

Foco nos sinais de feedback dos usuários, pegamos usuários que tem consumos similares e agrupa pra fazer a recomendação.

Uma característica histórica é que se tem um bom desempenho.

Possui uma baixa explicabilidade, não consigo explicar direito pq aquela recomendação está indo pro usuário.

Sofre bastante com a esparsidade, geralmente um usuário não consome todo seu catálogo e ai se cria uma matriz altamente esparsa, e os filtros colaborativos sofrem bastante com os efeitos dessas esparsidades. Mesmo assim ainda conseguem ter um desempenho satisfatório

Por fim, uma característica dos filtros colaborativos é uma maior serendipidade, o usuário receber algo que ele não esperava

image.png

4- Recomendação Híbrida que combina a saída de diferentes topologias. Posso pegar a saída de uma Recomendação Não Personalizada misturar com a Recomendação Baseada em Conteúdo e também com a saída de Filtragem Colaborativa

image.png

A Recomendação Híbrida tem uma maior complexidade de implementação, como se fosse um metamodelo, ou seja, vc tem um modelo em cima de outro modelo, geralmente se cria uma camada de complexidade alta pra se manter o algoritmo pra explicar a recomendação.

Por fim, vai ter mais hiperparâmetros pra se configurar, ou seja, vc vai ter o hiperparâmetros do seu mix de recomedação e também os parâmetros dos algortimos base que se usa pra fazer a recomendação híbrida.

Extrai um título da string e di qual o padrão de expressão regular 

```python
import re
def extract_year_from_title(title:str, regex='(\d{4})'):
    match = re.search(regex, title)
    return None if match is None else match.group()

def convert_genres_to_list(genres:str, separator='|'):
    return genres.split(separator)

df_items['year'] = df_items['title'].apply(extract_year_from_title)
df_items['genres_list'] = df_items['genres'].apply(convert_genres_to_list)
df_items.tail()
```

Essa função recebe uma série pandas como argumento em que o index é a categoria e os valores são os valores dessa categoria.

```python
def plot_category_count(category:pd.Series, figsize=(16, 8)):
    """Categorias como index"""
    fig, ax = plt.subplots(figsize=figsize)
    ax.barh(category.index, category.values)
    ax.grid(True, linestyle='--')
    ax.set_xlabel('Count')
    return fig, ax
```

.nunique() --> Quantos usuários distintos temos no dataset 

.shape[0] trás o número de linhas

.drop_duplicates() tirar valores duplicados

```python
n_users = df_users['user_id'].nunique()
n_items = df_items['item_id'].nunique()
n_ratings = df_ratings[['user_id', 'item_id']].drop_duplicates().shape[0]

print (f'#Usuários: {n_users}')
print (f'#Itens: {n_items}')
print (f'#Avaliações: {n_ratings}')
```

Aplicando a fórmula

```python
overall_sparsity = 1 - n_ratings / (n_users * n_items)
print ('Overall Sparsity: {:.02f} %'.format(100*overall_sparsity))
```

Vemos que Overall Sparsity (esparsidade geral) = 95.74 %, significa que temos aproximadamente 96% dos dados faltando no dataset

Queremos uma estrutura em que os usuários como linhas e os itens como colunas. 

.pivot --> transformar algumas linhas em colunas, precisamos passar quem serão as linhas ou index, itens serão as colunas e o valor que será colocado no usuários e itens.

.fillna(0) vai preecnher os valores nulos com zeros

```python
df_preferences = df_ratings.pivot(
    index='user_id', 
    columns='item_id',
    values='rating'
)
df_preferences = df_preferences.fillna(0)
df_preferences
```

image.png

.spy --> serve pra plotar uma matriz onde se o valor dessa matriz for maior do que 0 coloca um ponto preto, caso contrário coloca ponto branco.

```python
def plot_preference_matrix(df_preferences:pd.DataFrame, sparsity:float, figsize=(16,8)):
    fig, ax = plt.subplots(figsize=figsize)
    ax.spy(df_preferences.values.T)
    ax.set_ylabel('Items')
    ax.set_xlabel('Users')
    ax.set_title('Preference Matrix (Overall Sparsity {:.02f} %)'.format(100*sparsity))

    return fig, ax

plot_preference_matrix(df_preferences, overall_sparsity)
```

Vemos que a nossa esparsidade é de quase 96 %, vemos que temos muito spontos em branco, poucos pretos preenchidos.

```python
def plot_sparsity_distribution(sparsity, ylabel, figsize=(16,8), bins=100):
    fig, ax = plt.subplots(figsize=figsize)
    ax.grid(True, linestyle='--')
    ax.hist(100*sparsity, bins=bins)
    ax.set_ylabel(ylabel)
    ax.set_xlabel('Sparsity')
    ax.set_xlim(right=100.0)
    xticks = matplotlib.ticker.FormatStrFormatter('%.0f%%')
    ax.xaxis.set_major_formatter(xticks)
    return fig, ax
```

Vemos que muitos usuários tem uma esparsidade muito alta, isso pq muitos não avaliam

```python
plot_sparsity_distribution(users_sparsity, ylabel='Users')
```
